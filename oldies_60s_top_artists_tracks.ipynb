```python
filepath = "oldies_60s_top_artists_tracks.csv"
```

```python
df = pd.read_csv(filepath)
```

```python
df.head()
```

```python
for col in df.columns:
  if df[col].dtype == 'object':
    df[col] = df[col].str.lower()
```

```python
df['Duration_minutes'] = df['Duration (ms)'] / (1000 * 60)
df.drop(columns=['Duration (ms)'], inplace=True)
df.head()
```

```python
df['Duration_min'] = df['Duration_minutes'].round(2)
```

```python
df.head()
```

```python
df.drop(columns=['track id'], inplace=True)
```

```python
df.head()
```

```python
df['track_name'] = df['track name'].str.replace('-remastered 2009', '')
```
